```python
import math
import matplotlib.pyplot as plt


g = 9.81 # m / s^2
dt = 0.5
x_start = 0
y_start = 0
const = 4*(10**-5) # 1 / meter
v_initial = 700 # meters per second
angle_degree = [30,35,40,45,50,55]
angle = [math.radians(degrees) for degrees in angle_degree]
p_list = []
k = 0
for j in angle:

  vx = v_initial*math.cos(j)
  vy = v_initial*math.sin(j)

  x=[x_start]
  y=[y_start]

  i = 0

  while y[-1]>=0:
    x.append(x[i]+vx*dt)
    y.append(y[i]+vy*dt)
    vx=vx-const*dt*vx*math.sqrt((vx**2)+(vy**2))
    vy=vy-dt*(g+const*vy*math.sqrt((vx**2)+(vy**2)))
    i=i+1

  r = -y[-2]/y[-1]
  x[-1] = (x[-2]+r*x[-1])/(r+1)
  y[-1]=0.0

  plt.scatter(x,y, s=4)
  plt.xlabel('x (m)')
  plt.ylabel('y (m)')
  plt.title('Angle = {} Degrees'.format(angle_degree[k]))
  plt.show()
  k=k+1
```
